# Hilbert Spaces: From Euclidean Geometry to Infinite Dimensions

Hilbert spaces generalize familiar geometric ideas (distance, angles, projections) from 2D/3D to higher dimensions—even infinite ones (like functions). This is crucial for understanding linear models in forecasting and machine learning.

We'll start with 2D/4D examples, then show how the same math applies to functions!

## 1. Euclidean Space (2D) – Standard Geometry

Let's project a 2D point onto a line (subspace).

````python
import numpy as np
import matplotlib.pyplot as plt

# Define a vector (point) and a line (subspace)
v = np.array([3, 2])  # Original point
u = np.array([1, 1])  # Direction of the line

# Projection of v onto u
proj = (np.dot(v, u) / np.dot(u, u)) * u

# Plot
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='Original Vector (v)')
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b', label='Subspace (u)')
plt.quiver(0, 0, proj[0], proj[1], angles='xy', scale_units='xy', scale=1, color='g', label='Projection (proj)')
plt.plot([v[0], proj[0]], [v[1], proj[1]], 'k--', label='Error (orthogonal)')
plt.xlim(0, 4)
plt.ylim(0, 3)
plt.grid()
plt.legend()
plt.title('2D Projection: v onto u')
plt.show()
````

**Key Idea:**
- The green vector is the "shadow" of $v$ on $u$.
- The dashed line is the error (orthogonal to $u$).
- This is how least-squares regression works in 2D!

## 2. Higher Dimensions (4D) – Still Euclidean, but Hard to Visualize

In 4D, we can't plot easily, but the math is the same.

````python
# 4D vectors
v_4d = np.array([1, 2, 3, 4])
u_4d = np.array([0, 1, 0, 1])  # Subspace direction

# Projection (same formula as 2D!)
proj_4d = (np.dot(v_4d, u_4d) / np.dot(u_4d, u_4d)) * u_4d

print("Original 4D vector:", v_4d)
print("Projection onto u_4d:", proj_4d)
````

**Output:**
```
Original 4D vector: [1 2 3 4]
Projection onto u_4d: [0. 3. 0. 3.]
```
**Key Idea:**
- The projection formula generalizes to any dimension (even 1000D!).
- Hilbert spaces extend this to infinite dimensions (like functions).

## 3. Infinite-Dimensional Hilbert Space (Functions!)

Now, imagine a function (e.g., a time series $f(t)$). We can still project it onto a subspace (like a set of polynomials)!

````python
from numpy.polynomial import Polynomial

# Define a function f(t) = t^2 + noise
t = np.linspace(0, 1, 100)
np.random.seed(0)
f = t**2 + 0.1 * np.random.randn(100)

# Fit a linear polynomial (subspace of degree-1 polynomials)
p = Polynomial.fit(t, f, deg=1)

# Plot
plt.figure(figsize=(8, 4))
plt.scatter(t, f, s=5, label='Original f(t) (noisy)')
plt.plot(t, p(t), 'r', label='Projection onto degree-1 polynomials')
plt.title('Projecting a Function onto a Subspace')
plt.legend()
plt.grid()
plt.show()
````

**Key Idea:**
- The red line is the "shadow" (projection) of the noisy $f(t)$ onto the subspace of linear functions.
- This is how linear regression works for time series forecasting!

## Why Hilbert Spaces Matter for Forecasting

1. **Linear Models** (e.g., ARIMA, Linear Regression):
   - Use projections to find the "best-fit" line/curve in Hilbert space.
   - Example: Fitting a trendline to stock prices.

2. **Non-Linear Models (GBDT, Neural Nets):**
   - Don’t rely on projections; instead, they iteratively adjust to minimize error.
   - Example: A decision tree splits data like a flowchart, not using "shadows."

## Summary

| Concept          | 2D/3D (Euclidean) | Higher-D (Hilbert) | Forecasting Use Case          |
|------------------|-------------------|--------------------|-------------------------------|
| **Vectors**      | `[1, 2]`          | `[1, 2, 3, ...]`   | Features in a dataset         |
| **Projections**  | Shadow on a line  | Best-fit function  | Linear regression             |
| **Inner Product**| Dot product       | Integrals (for functions) | Measuring similarity |

**Hilbert spaces let us generalize "shadows" (projections) to functions, which is why they’re key for linear forecasting models!**
